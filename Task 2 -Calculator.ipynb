```python
#CALUCULATOR
import tkinter as tk
from tkinter import messagebox

# Function to perform the calculation
def calculate():
    try:
        num1 = float(entry_num1.get())
        num2 = float(entry_num2.get())
        operation = operator_var.get()

        if operation == "+":
            result.set(num1 + num2)
        elif operation == "-":
            result.set(num1 - num2)
        elif operation == "*":
            result.set(num1 * num2)
        elif operation == "/":
            if num2 != 0:
                result.set(num1 / num2)
            else:
                result.set("Error: Division by zero")
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter valid numbers.")
    except Exception as e:
        messagebox.showerror("Error", str(e))

# Create the main application window
app = tk.Tk()
app.title("Simple Calculator")

# Create entry fields for input
entry_num1 = tk.Entry(app)
entry_num2 = tk.Entry(app)
entry_num1.grid(row=0, column=1)
entry_num2.grid(row=0, column=3)

# Create labels and entry fields for number1 and number2
num1_label = tk.Label(app, text="Number 1:")
num2_label = tk.Label(app, text="Number 2:")
num1_label.grid(row=0, column=0)
num2_label.grid(row=0, column=2)

# Create a label for the operator
operator_var = tk.StringVar()
operator_var.set("+")  # Default operator
operator_label = tk.Label(app, text="Operator:")
operator_label.grid(row=1, column=0)
operator_dropdown = tk.OptionMenu(app, operator_var, "+", "-", "*", "/")
operator_dropdown.grid(row=1, column=1, columnspan=3)

# Create a button to calculate the result
calculate_button = tk.Button(app, text="Calculate", command=calculate)
calculate_button.grid(row=2, column=0, columnspan=4)

# Create a label to display the result
result = tk.StringVar()
result_label = tk.Label(app, text="Result:")
result_label.grid(row=3, column=0, columnspan=4)
result_display = tk.Label(app, textvariable=result)
result_display.grid(row=4, column=0, columnspan=4)

# Start the GUI main loop
app.mainloop()
```
