```python
#TO-DO-LIST
import tkinter as tk
from tkinter import messagebox

# Create a function to add a new task
def add_task():
    task = task_entry.get()
    if task:
        task_list.insert(tk.END, task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

# Create a function to mark a task as complete
def mark_completed():
    selected_task = task_list.curselection()
    if selected_task:
        task_index = selected_task[0]
        task_list.itemconfig(task_index, {'bg':'green'})
        completed_listbox.insert(tk.END, task_list.get(task_index))
        task_list.delete(task_index)
    else:
        messagebox.showwarning("Warning", "Please select a task to mark as completed.")

# Create a function to delete a task
def delete_task():
    selected_task = task_list.curselection()
    if selected_task:
        task_index = selected_task[0]
        task_list.delete(task_index)
    else:
        messagebox.showwarning("Warning", "Please select a task to delete.")

# Create the main application window
app = tk.Tk()
app.title("To-Do List")

# Create an entry field to add tasks
task_entry = tk.Entry(app, width=40)
task_entry.pack(pady=10)

# Create a button to add tasks
add_button = tk.Button(app, text="Add Task", command=add_task)
add_button.pack()

# Create a listbox to display tasks
task_list = tk.Listbox(app, selectmode=tk.SINGLE, width=40)
task_list.pack(pady=10)

# Create a button to mark tasks as complete
mark_button = tk.Button(app, text="Mark as Completed", command=mark_completed)
mark_button.pack()

# Create a button to delete tasks
delete_button = tk.Button(app, text="Delete Task", command=delete_task)
delete_button.pack()

# Create a separate listbox to display completed tasks
completed_listbox = tk.Listbox(app, selectmode=tk.SINGLE, width=40, bg='green', fg='white')
completed_listbox.pack(pady=10)

# Start the GUI main loop
app.mainloop()
```
