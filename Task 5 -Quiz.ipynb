```python
#QUIZ
import tkinter as tk
from tkinter import messagebox
import random

# Define quiz questions and answers
questions = [
    {
        "question": "What is the capital of France?",
        "options": ["Paris", "London", "Berlin", "Madrid"],
        "correct_answer": "Paris"
    },
    {
        "question": "What is the largest planet in our solar system?",
        "options": ["Earth", "Mars", "Jupiter", "Venus"],
        "correct_answer": "Jupiter"
    },
    {
        "question": "What is 2 + 2?",
        "options": ["3", "4", "5", "6"],
        "correct_answer": "4"
    }
]

# Initialize variables
current_question_index = 0
score = 0

# Create a function to check the answer
def check_answer():
    global current_question_index
    global score
    selected_option = var.get()
    correct_answer = questions[current_question_index]["correct_answer"]
    
    if selected_option == correct_answer:
        score += 1

    current_question_index += 1
    if current_question_index < len(questions):
        load_question()
    else:
        show_final_results()

# Create a function to load the next question
def load_question():
    question_label.config(text=questions[current_question_index]["question"])
    random.shuffle(questions[current_question_index]["options"])
    
    for i in range(4):
        option_buttons[i].config(text=questions[current_question_index]["options"][i], value=questions[current_question_index]["options"][i])

# Create a function to display the final results
def show_final_results():
    result_message = f"You scored {score}/{len(questions)}!"
    messagebox.showinfo("Quiz Complete", result_message)
    root.destroy()

# Create the main GUI window
root = tk.Tk()
root.title("Quiz Game")

# Create and configure widgets
question_label = tk.Label(root, text="", font=("Helvetica", 14))
question_label.pack(pady=20)

var = tk.StringVar()

option_buttons = []
for i in range(4):
    option_button = tk.Radiobutton(root, text="", variable=var, font=("Helvetica", 12))
    option_buttons.append(option_button)
    option_button.pack()

next_button = tk.Button(root, text="Next", command=check_answer, font=("Helvetica", 12))
next_button.pack(pady=20)

# Load the first question
load_question()

# Start the GUI main loop
root.mainloop()
```
