```python
#PASSWORD GENERATOR
import tkinter as tk
import random
import string

# Function to generate a random password
def generate_password():
    password_length_str = length_entry.get()
    
    # Check if the input is a valid integer
    if not password_length_str.isdigit():
        result_label.config(text="Invalid input. Please enter a valid integer.")
        return

    password_length = int(password_length_str)
    
    # Define characters for password generation
    characters = string.ascii_letters + string.digits + string.punctuation
    
    # Generate a random password
    password = ''.join(random.choice(characters) for _ in range(password_length))
    
    # Clear previous result and display the generated password
    result_label.config(text=f"Generated Password for {username_entry.get()}: {password}")

# Function to reset the generated password
def reset_password():
    length_entry.delete(0, tk.END)
    result_label.config(text="")
    username_entry.delete(0, tk.END)

# Create the main GUI window
root = tk.Tk()
root.title("Password Generator")

# Create and configure widgets
username_label = tk.Label(root, text="Enter Username:", font=("Helvetica", 14))
username_label.pack(pady=10)

username_entry = tk.Entry(root, font=("Helvetica", 12))
username_entry.pack(pady=5)

length_label = tk.Label(root, text="Enter Password Length:", font=("Helvetica", 14))
length_label.pack(pady=10)

length_entry = tk.Entry(root, font=("Helvetica", 12))
length_entry.pack(pady=5)

generate_button = tk.Button(root, text="Generate Password", command=generate_password, font=("Helvetica", 12))
generate_button.pack(pady=10)

reset_button = tk.Button(root, text="Reset", command=reset_password, font=("Helvetica", 12))
reset_button.pack(pady=5)

result_label = tk.Label(root, text="", font=("Helvetica", 12))
result_label.pack()

# Start the GUI main loop
root.mainloop()
```
